„Duomenų atstatymas, taikant skirtingų metrikų skaičiavimą“. Parašykite programą, kuri atstatytų (1 lentelės) praleistą reikšmę, taikant skirtingų metrikų skaičiavimą, t. y. Euklido, Manhattan ir Čebyševo metriką. Atlikite duomenų normalizavimą, naudojant vidurkį ir vidutinį kvadratinį nuokrypį.

```python
import pandas as pd
import numpy as np
from scipy.spatial import distance

lentele2 = lentele.copy()
lentele2['Ûgis'] = lentele2['Ûgis'].replace('?', pd.NA)
lentele2['Ûgis'] = pd.to_numeric(lentele2['Ûgis'], errors='coerce')
# lentele2 = lentele2[lentele2.Lytis == 1]
lentele2
```

```python
ugis_vidurkis = lentele2['Ûgis'].mean()
ugis_std = lentele2['Ûgis'].std()

svoris_vidurkis = lentele2['Svoris'].mean()
svoris_std = lentele2['Svoris'].std()

lytis_vidurkis = lentele2['Lytis'].mean()
lytis_std = lentele2['Lytis'].std()

print('ugis_vidurkis:', ugis_vidurkis, 'ugis_std:',  ugis_std)
print('svoris_vidurkis:', svoris_vidurkis, 'svoris_std:',  svoris_std)
print('lytis_vidurkis:', lytis_vidurkis, 'lytis_std:',  lytis_std)
```

```python
# Duomenu normalizacija
normal_ugis = (lentele2['Ûgis'] - ugis_vidurkis) / ugis_std
normal_svoris = (lentele2['Svoris'] - svoris_vidurkis) / svoris_std
normal_lytis = (lentele2['Lytis'] - lentele2['Lytis'].mean()) / lentele2['Lytis'].std()
print([[normal_svoris], [normal_ugis], [normal_lytis]])
print(normal_svoris[4])
```

```python
ats_euklidas = distance.euclidean((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_euklidas:', ats_euklidas)

ats_manhatan = distance.cityblock((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_manhatan:', ats_manhatan)

ats_cebysev = distance.chebyshev((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_cebysev:', ats_cebysev)
```

```python
# ats_euklidas - Atstumų masyvas (Euclidean Distances):
# kintamasis, kuriame saugomi Euklido atstumai tarp ketvirto asmens ir visų kitų asmenų lentelėje.
print(normal_lytis[4], normal_svoris[4])
ats_euklidas = np.sqrt((normal_lytis[4] - normal_lytis)**2 + (normal_svoris[4] - normal_svoris)**2)
# print(ats_euklidas)
del ats_euklidas[4]
print('ats_euklidas:\n', ats_euklidas)

ats_manhatan = np.abs(normal_lytis[4] - normal_lytis) + np.abs(normal_svoris[4] - normal_svoris)
del ats_manhatan[4]
print('ats_manhatan:\n', ats_manhatan)

ats_cebysev = max(normal_lytis[4] - normal_lytis) + np.abs(normal_svoris[4] - normal_svoris)
del ats_cebysev[4]
print('ats_cebysev:\n', ats_cebysev)
```

```python
# ats_euklidas = 1.398039
```

```python
# normalizuotas_ugis_eukl: kintamasis, kuriame saugomas normalizuotas svorio atstumas (pagal Euklido atstumą).
normalizuotas_ugis_eukl = (1 / np.sum(1 / ats_euklidas) * np.sum(normal_ugis / ats_euklidas))
print("normalizuotas_ugis_eukl", normalizuotas_ugis_eukl)

# ugis_eukl: kintamasis, kuriame saugomas galutinis svorio vidurkis, apskaičiuotas naudojant Euklido atstumo normalizaciją.
ugis_eukl = normalizuotas_ugis_eukl * ugis_std + ugis_vidurkis
print('Vidurkis pagal Euklida:', ugis_eukl)

# ----------------------------------------------------------------------------
normalizuotas_ugis_city = (1 / np.sum(1 / ats_manhatan) * np.sum(normal_ugis / ats_manhatan))
print("normalizuotas_ugis_city:", normalizuotas_ugis_city)

ugis_city = normalizuotas_ugis_city * ugis_std + ugis_vidurkis
print('Vidurkis pagal Manhatan:', ugis_city)

# ----------------------------------------------------------------------------
normalizuotas_ugis_cheb = (1 / np.sum(1 / ats_cebysev) * np.sum(normal_ugis / ats_cebysev))
print("normalizuotas_ugis_cheb:", normalizuotas_ugis_cheb)

ugis_cheb = normalizuotas_ugis_cheb * ugis_std + ugis_vidurkis
print('Vidurkis pagal Chebyshev:', ugis_cheb)
```

Neatitikčių (anomalijų) paieška Kiekviena mokykla pateikia Lietuvos olimpiniam komitetui dalyvių sąrašą sporto varžyboms. Viena iš jų pateikė sąrašą, kuriame buvo nurodytas toks vaikų amžius, t. y. duomenų masyvas: A=[15 17 16 16 18 17 15 30 18].  

6 uždavinys „Neatitikčių paieška, taikant Šovene kriterijų“. Duotajame A masyve raskite neatitikčių (anomalijų), taikant Šovene kriterijų.

Šovene kriterijus, tai yra, kai elementas laikomas išskirtimi, jei jo nuokrypis nuo vidurkio yra didesnis nei tam tikras daugiklis (pavyzdžiui, 2) kartu su standartiniu nuokrypiu.
Dabartinio atsakymo pavyzdžiui:
Apskaičiuokite vidurkį:
Vidurkis = (15 + 17 + 16 + 16 + 18 + 17 + 15 + 30 + 18) / 9 = 162 / 9 = 18.
Apskaičiuokite standartinį nuokrypį (SD) nuo vidurkio. Standartinis nuokrypis yra skirtumas tarp kiekvieno elemento ir vidurkio kvadratu, iš visų elementų kvadratų vidurkio šaknies:
SD = √[((15 - 18)² + (17 - 18)² + (16 - 18)² + (16 - 18)² + (18 - 18)² + (17 - 18)² + (15 - 18)² + (30 - 18)² + (18 - 18)²) / 9]
SD = √[((9 + 1 + 4 + 4 + 0 + 1 + 9 + 144 + 0) / 9)]
SD = √[(172 / 9)]
SD ≈ √19 ≈ 4.36.
Dabar, naudodami Šoveno kriterijų, galime nustatyti išskirtis, kurios yra didesnės nei 2 kartus standartinis nuokrypis nuo vidurkio:
Išskirtis = 2 * SD ≈ 2 * 4.36 ≈ 8.72.

```python
import numpy as np
from scipy import stats

# Duomenų masyvas
A = [15, 17, 16, 16, 18, 17, 15, 30, 18]

# Apskaičiuojame vidurkį ir standartinį nuokrypį
mean = np.mean(A)
std_dev = np.std(A)

# Nustatome slenkstį, daugiklis nuo 2 iki 3
threshold = 2 * std_dev  # Arba 3 * std_dev, priklauso nuo reikalavimų

# Ieškome išskirtų reikšmių
outliers = [x for x in A if abs(x - mean) > threshold]

print("Duomenų masyvas:", A)
print("Vidurkis:", mean)
print("Standartinis nuokrypis:", std_dev)
print("Išskirtys:", outliers)
```

# 7 uždavinys


---
Neatitiktimis dažnai yra laikomos reikšmės, kurios yra žymiai mažesnės nei 1-asis kvartilis ar žymiai didesnės nei 3-asis kvartilis.
Jeigu elementas nepriklauso intervalui.𝑄25−1.5∙(𝑄75-𝑄25), 𝑄75+1.5∙(𝑄75-𝑄25)
7 uždavinys Duotajame A masyve raskite neatitikčių (anomalijų), taikant Kvartilių algoritmą.

```python
import numpy as np

# Duomenų masyvas
A = [15, 17, 16, 16, 18, 17, 15, 30, 18, 51]

# Apskaičiuojame kvartilius
Q25 = np.percentile(A, 25)
Q75 = np.percentile(A, 75)

# Apskaičiuojame interkvartilių intervalą (IQR)
IQR = Q75 - Q25

# Nustatome slenksčius anomalijoms
lower_threshold = Q25 - 1.5 * IQR
upper_threshold = Q75 + 1.5 * IQR

# Ieškome anomalijų
outliers = [x for x in A if x < lower_threshold or x > upper_threshold]

print("Duomenų masyvas:", A)
print("Pirmasis kvartilis (Q25):", Q25)
print("Trečiasis kvartilis (Q75):", Q75)
print("Interkvartilių intervalas (IQR):", IQR)
print("Anomalijos:", outliers)
```

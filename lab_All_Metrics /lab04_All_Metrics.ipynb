```python
import pandas as pd
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
lentele = pd.read_csv("*.csv", encoding="ISO-8859-1", header=0)
lentele
```

```python
# pakeičiu simbolius „?“ į trūkstamas reikšmes, konvertuoju visus „Ûgis“ stulpelio įrašus į skaičius, ir išrenku eilutes pagal stulpelio „Lytis“ reikšmę 1
lentele['Ûgis'] = lentele['Ûgis'].replace('?', pd.NA)
lentele['Ûgis'] = pd.to_numeric(lentele['Ûgis'], errors='coerce')
# lentele = lentele[lentele.Lytis == 1]
lentele
```

```python
ugis_vidurkis = lentele['Ûgis'].mean()
ugis_std = lentele['Ûgis'].std()

svoris_vidurkis = lentele['Svoris'].mean()
svoris_std = lentele['Svoris'].std()

lytis_vidurkis = lentele['Lytis'].mean()
lytis_std = lentele['Lytis'].std()

print('ugis_vidurkis:', ugis_vidurkis, 'ugis_std:',  ugis_std)
print('svoris_vidurkis:', svoris_vidurkis, 'svoris_std:',  svoris_std)
print('lytis_vidurkis:', lytis_vidurkis, 'lytis_std:',  lytis_std)
```

```python
# skaičiuojam vidurkius ir standartinius nuokrypius stulpeliams „Ûgis“, „Svoris“ ir „Lytis“, taip paruošdami duomenis tolesnei statistinei analizei arba normalizacijai.
normal_ugis = (lentele['Ûgis'] - ugis_vidurkis) / ugis_std
normal_svoris = (lentele['Svoris'] - svoris_vidurkis) / svoris_std
normal_lytis = (lentele['Lytis'] - lytis_vidurkis) / lytis_std

print([[normal_svoris], [normal_ugis], [normal_lytis]])
print(normal_svoris[4])
```

```python
ats_euklidas = distance.euclidean((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_euklidas:', ats_euklidas)

ats_manhatan = distance.cityblock((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_manhatan:', ats_manhatan)

ats_cebysev = distance.chebyshev((normal_lytis[4]-normal_lytis), (normal_svoris[4]-normal_svoris))
print('ats_cebysev:', ats_cebysev)
```

```python
# ats_euklidas - Atstumų masyvas (Euclidean Distances):
# kintamasis, kuriame saugomi Euklido atstumai tarp ketvirto asmens ir visų kitų asmenų lentelėje.
print(normal_lytis[4], normal_svoris[4])
ats_euklidas = np.sqrt((normal_lytis[4] - normal_lytis)**2 + (normal_svoris[4] - normal_svoris)**2)
# print(ats_euklidas)
del ats_euklidas[4]
print('ats_euklidas:\n', ats_euklidas)

ats_manhatan = np.abs(normal_lytis[4] - normal_lytis) + np.abs(normal_svoris[4] - normal_svoris)
del ats_manhatan[4]
print('ats_manhatan:\n', ats_manhatan)

ats_cebysev = max(normal_lytis[4] - normal_lytis) + np.abs(normal_svoris[4] - normal_svoris)
del ats_cebysev[4]
print('ats_cebysev:\n', ats_cebysev)

# Skirtingų metrikų (Euklido, Manhattan ir Čebyševo) skaičiuotus atstumus tarp pasirinkto
# asmens (4-as asmuo, kuris yra išimtas iš sąrašo, nes jis yra taškas, nuo kurio skaičiuojami atstumai) ir kitų asmenų.
```

```python
import pandas as pd
from scipy.spatial import distance

lentele = pd.read_csv('*.csv', header = 0)
rekomendac_lentele = lentele.copy()
rekomendac_lentele
```

```python
# Vartotojo įvesties gavimas - pasirinkimas prekei iš lentelės pagal jos eilutės indeksą
pasirinktas_indeksas = int(input("Įveskite prekės eilutės indeksą: ")) - 1  # -1, nes sąrašo indeksai prasideda nuo 0
print('pasirinktas_indeksas:', pasirinktas_indeksas)
```

```python
if 0 <= pasirinktas_indeksas < len(lentele):
    pasirinkta_preke = lentele.iloc[pasirinktas_indeksas, 1:].values  # Pasirinktos prekės vektorius be pirmosios reikšmės
    # print(pasirinkta_preke)

    atstumai = []  # atstumų rezultatai

    # Skaičiuojame Euklido atstumą tarp pasirinktos prekės ir visų kitų prekių
    for indeksas, eilute in lentele.iterrows():
        kandidato_preke = eilute[1:].values
        # print('kandidat preke:', kandidato_preke)
        atstumas = distance.euclidean(pasirinkta_preke, kandidato_preke)
        # print('atstumas:',atstumas)
        # print('indeksas: ', indeksas)
        # print('eilute: ', eilute)
        atstumai.append((lentele.iloc[indeksas, 0], atstumas))  # Saugome prekės pavadinimą ir atstumą
        # print('atstumai: ', atstumai)

    # Rikiuojame prekes pagal atstumą
    atstumai.sort(key=lambda x: x[1])

    # Spausdiname visus atstumus
    for preke, atstumas in atstumai:
        print(f"Atstumas tarp {lentele.iloc[pasirinktas_indeksas, 0]} ir '{preke}'bus: {atstumas}")


    # Rekomenduojama prekė - artimiausia
    rekomenduota_preke = atstumai[1][0]
    # Spausdiname rekomenduojamą prekę
    print(f"Rekomenduojama prekė: {rekomenduota_preke}")

else:
    print("Neteisingas indeksas. Įveskite galiojantį indeksą.")
```

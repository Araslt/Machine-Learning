```python
import pandas as pd

# Susikuriam zodyna su vartotojų įvertinimais
data = {
    'Filmai / Žanrai': ['Komedija', 'Drama', 'Veiksmo', 'Fantastika', 'Siaubo', 'Romantika', 'Trileriai'],
    'Vartotojas 1': [3, 4, 2, 1, 5, 2, 3],
    'Vartotojas 2': [4, 5, 3, 2, 1, 4, 3],
    'Vartotojas 3': [2, 3, 4, 5, 1, 2, 4],
    'Vartotojas 4': [5, 4, 3, 3, 2, 1, 2],
    'Vartotojas 5': [1, 2, 5, 4, 3, 3, 5],
    'Vartotojas 6': [5, 4, 3, 4, 2, None, None]
}
```

```python
# Lentele transponuojam, kad vartotojai būtų eilutėse, o žanrai - stulpeliuose
df_lentele = pd.DataFrame(data).set_index('Filmai / Žanrai').T
df_lentele
```

```python
df_lentele.loc['Vartotojas 6']
```

```python
# Paimamm šeštąją eilutę, ir palyginam ją su visomis kitomis eilutėmis
# tuo budu rasdami koreliacijos koeficientus tarp Vartotojo 6 ir kitų vartotojų skonių
# rezultatas yra pandas Series!!!
correlations = df_lentele.corrwith(df_lentele.loc['Vartotojas 6'], axis=1)
correlations
```

```python
# Pašalinam 6-ąjį vartotoją iš koreliacijų
correlations = correlations.drop('Vartotojas 6')
correlations
```

```python
# Randame vartotoją su didžiausia koreliacija (panašiausiu skoniu)
# tai atliekame su .idxmax() yra pandas Series metodu, kuris naudojamas rasti indeksą (vartotoją),
# kurio koreliacijos koeficientas yra didžiausias.
most_similar_user = correlations.idxmax()
print("most_similar_user - ", most_similar_user)
df_lentele
```

```python
# Rekomenduokite šeštajam vartotojui "Romantikos" ir "Trilerių" žanrų reikšmes iš panašiausio vartotojo
recommended_romantic = df_lentele.at[most_similar_user, 'Romantika']
recommended_thriller = df_lentele.at[most_similar_user, 'Trileriai']

print(f"Rekomenduojamos 'Romantikos' ir 'Trilerių' žanrų reikšmės šeštajam vartotojui:")
print(f"Romantika: {recommended_romantic}")
print(f"Trileriai: {recommended_thriller}")
```
